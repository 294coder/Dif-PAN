```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 20
fontdict = {'family':'Times New Roman', 'style':'italic'}

plt.figure(figsize=(8*1.9,6))
res = [64, 128, 256, 512, 1024]

conv_mem = np.array([7.4, 197, 410, 1190, 4.340*1e3]) / 1e3
conv_flops = [0.439, 1.76, 7.03, 28.12, 112.5]

mamba_mem = np.array([14, 278, 764, 2.9*1e3, 11.539*1e3]) / 1e3
mamba_flops = [2.48, 9.836, 39.246, 156.9, 627.4]

attn_mem = np.array([28, 513, 12.9e3, 78e3]) / 1e3
attn_flops = np.array([5.89, 25.36, 178.3, 582.1])

swin_mem = np.array([12.9, 492.1, 1498, 9.391e3, 38.2e3]) / 1e3
swin_flops = np.array([3.29, 13.16, 78.92, 292.1, 1032.3])

plt.subplot(1,2,1)
plt.plot(res, conv_mem, marker='^', linestyle='--', lw=2, markersize=10, label='Conv')
plt.plot(res[:4], attn_mem, marker='*', linestyle='--', lw=2, markersize=12, label='Attention')
plt.plot(res, swin_mem, marker='o', linestyle='--', lw=2, markersize=10, label='Swin')
plt.plot(res, mamba_mem, marker='v', lw=2, markersize=10, label='LEMM')

plt.grid(linestyle='--')
plt.xticks([64, 128, 256, 512, 1024])
plt.xlabel('Resolution (px)')
plt.ylabel('GPU Memory (GB)')
plt.ylim([-1, 60])
plt.legend()

plt.subplot(1,2,2)
plt.plot(res, conv_flops, marker='^', linestyle='--', lw=2, markersize=10, label='Conv')
plt.plot(res[:4], attn_flops, marker='*', linestyle='--', lw=2, markersize=12, label='Attention')
plt.plot(res, swin_flops, marker='o', linestyle='--', lw=2, markersize=10, label='Swin')
plt.plot(res, mamba_flops, marker='v', lw=2, markersize=10, label='LEMM')

plt.grid(linestyle='--')
plt.xticks([64, 128, 256, 512, 1024])
plt.xlabel('Resolution (px)')
plt.ylabel('FLOPs (GB)')
plt.legend()
plt.gca().yaxis.set_label_coords(-0.08, 0.5)
plt.tight_layout()

# plt.subplots_adjust()

plt.savefig('mem_flops.png', dpi=300)
```
